```python
import pandas as pd
import re
```

```python
file_path = 'Market Basket Analysis 5 - Market Basket Analysis 5.csv.csv'
with open(file_path, 'r') as file:
    lines = file.readlines()
```

```python
# Split Date and Item
def split_date_item_improved(row):
  match = re.match(r'(\d{1,2}/\d{1,2}/\d{4})(.*)', row)
  if match:
      date = match.group(1)
      items = match.group(2)
      return date, items
  else:
      return None, None

# Count items
def count_items(item_str):
  items = item_str.split(',')
  return len([item for item in items if item.strip()])


def preprocessing(lines):
  # Split Rows
  split_data = [split_date_item_improved(row) for row in lines]
  split_df = pd.DataFrame(split_data, columns=['Date', 'Item'])

  split_df['Total Items'] = split_df['Item'].apply(count_items)

  return split_df
```

```python
data = preprocessing(lines)
data
```

```python
df = data['Item'].str.split(',', expand=True)
# binary_df = data['Item'].str.join('|').str.get_dummies()
```

```python
import pandas as pd
from pyECLAT import ECLAT

# Apply the Eclat algorithm to find frequent itemsets
min_support = 0.4  # You can adjust this threshold as needed
frequent_itemsets = ECLAT(df, verbose=True)

```

```python
frequent_itemsets.df_bin
```

```python
frequent_itemsets.uniq_
```

```python
get_ECLAT_indexes, get_ECLAT_supports = frequent_itemsets.fit(min_support=0.3,
                                                           min_combination=1,
                                                           max_combination=3,
                                                           separator=' & ',
                                                           verbose=True)
```
